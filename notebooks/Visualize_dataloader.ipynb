```python
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from probunet.data_utils.dataloader import BratsDataset

%matplotlib inline
```

```python
dataset = BratsDataset('<TRAIN_PATH>')
dataloader = DataLoader(dataset, shuffle=True, batch_size=8, collate_fn=dataset.collate_batch)
```

```python
X, y = next(iter(dataloader))
```

```python
for batch_ix in range(8):
    
    fig, ax = plt.subplots(nrows=1, ncols=7, figsize=(10, 10))
    
    for input_ix, input_title in enumerate(["T1", "T2", "T1C", "FLAIR"]):
        cl_slice = X[batch_ix, input_ix, :, :].numpy()
        ax[input_ix].imshow(cl_slice, cmap='gray')
        ax[input_ix].set_title(input_title)
        ax[input_ix].axis('off')
        
    for y_ix, y_title in enumerate(["WT", "TC", "ET"]):
        y_slice = y[batch_ix, y_ix, :, :].numpy()
        ax[4+y_ix].imshow(y_slice, cmap='gray')
        ax[4+y_ix].set_title(y_title)
        ax[4+y_ix].axis('off')
    
    plt.show()
```
